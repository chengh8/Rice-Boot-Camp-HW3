### Heroes of Pymoli Data Analysis

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

### PLAYER COUNT
* Total Number of Players

```python
Unique_Players = purchase_df["SN"].unique()
Total_dic = {"Total Players":[len(Unique_Players)]}
Total_Players = pd.DataFrame(Total_dic)
Total_Players
```

### PURCHASING ANALYSIS (TOTAL)
* Number of Unique Items
* Average Purchase Price
* Total Number of Purchases
* Total Revenue

```python
Unique_Items = purchase_df["Item ID"].nunique()
Avg_Price = purchase_df["Price"].sum()/len(purchase_df["Price"])
Total_Purch = purchase_df["Purchase ID"].count()
Total_Rev = purchase_df["Price"].sum()

PurchAnaly_dic = {"Number of Unique Items":[Unique_Items],
                  "Average Purchase Price":[Avg_Price],
                  "Total Number of Purchases":[Total_Purch],
                  "Total Revenue":[Total_Rev]                                           
                }
PurchAnaly_df = pd.DataFrame(PurchAnaly_dic)
PurchAnaly_df["Average Purchase Price"] = PurchAnaly_df["Average Purchase Price"].map("${:,.2f}".format)
PurchAnaly_df["Total Revenue"] = PurchAnaly_df["Total Revenue"].map("${:,.2f}".format)
PurchAnaly_df
```

## Gender Demographics
* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
Gender_df = purchase_df.drop_duplicates(subset=["SN"], keep="first")
Gender_counts = Gender_df["Gender"].value_counts()
Gender_summary_table = pd.DataFrame({"Purchase ID": Gender_counts})
Gender_summary_table["Percentage of Players"] = (Gender_counts/len(Unique_Players))*100
Gender_summary_table["Percentage of Players"] = Gender_summary_table["Percentage of Players"].map("{:,.2f}".format)
Gender_summary_table
```

## Purchasing Analysis (Gender)
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
Gender_Purch = purchase_df["Gender"].value_counts()
Gender_group = purchase_df.groupby(["Gender"])
Total_Price = Gender_group["Price"].sum()
Avg_Price = Total_Price/Gender_Purch
Avg_Per = purchase_df.groupby(["Gender"])["Price"].sum()/Gender_counts
Gender_Price_table = pd.DataFrame({"Purchase Count": Gender_Purch,
                                     "Average Purchase Price": Avg_Price,
                                     "Total Purchase Price": Total_Price,
                                     "Avg Total Purchase per Person":Avg_Per
                                     })
Gender_Price_table["Average Purchase Price"] = Gender_Price_table["Average Purchase Price"].map("${:,.2f}".format)
Gender_Price_table["Avg Total Purchase per Person"] = Gender_Price_table["Avg Total Purchase per Person"].map("${:,.2f}".format)
Gender_Price_table
```

## Age Demographics
* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table

```python
Age_Bins_df = purchase_df.drop_duplicates(subset=["SN"], keep="first")
Bins = [0,9,14,19,24,29,34,39,50]
Age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
Age_Bins_df["Age_Grp"] = pd.cut(Age_Bins_df["Age"], Bins, labels=Age_labels)
Age_group = Age_Bins_df.groupby("Age_Grp")

Age_table = pd.DataFrame()
Age_table["Total Count"] = Age_group["Age"].count()
Age_table["Percentage of Players"] = (Age_table["Total Count"]/len(Unique_Players))*100
Age_table["Percentage of Players"] = Age_table["Percentage of Players"].map("{:.2f}".format)
Age_table
```

## Purchasing Analysis (Age)
* Bin the purchase_data data frame by age
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
#This was to create a copy of purchase_df
Age_Purch_df = purchase_df

#Put the ages in bins again with duplicates
Age_Purch_df["Age_Grp"] = pd.cut(purchase_df["Age"], Bins, labels=Age_labels)
Age_Uni_df = Age_Purch_df.drop_duplicates(subset=["SN"], keep="first")
Group_Age = Age_Purch_df.groupby("Age_Grp")
Age_Grp_count = Age_Uni_df["Age_Grp"].value_counts()
TotalPrice = Group_Age["Price"].sum()

#Creates an empty Dataframe
Total_AgePurch_df = pd.DataFrame()
Total_AgePurch_df["Purchase Count"] = Group_Age["Purchase ID"].count()
Total_AgePurch_df["Average Purchase Price"] = TotalPrice/Total_AgePurch_df["Purchase Count"]
Total_AgePurch_df["Total Purchase Value"] = TotalPrice
Total_AgePurch_df["Avg Total Purchase per Person"] = Age_Purch_df.groupby(["Age_Grp"])["Price"].sum()/Age_Grp_count

Total_AgePurch_df["Average Purchase Price"] = Total_AgePurch_df["Average Purchase Price"].map("${:,.2f}".format)
Total_AgePurch_df["Avg Total Purchase per Person"] = Total_AgePurch_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
Total_AgePurch_df
```

## Top Spenders
* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame

```python
Spenders = purchase_df.groupby("SN")
Top_Spenders = pd.DataFrame()
Top_Spenders["Purchase Count"] = Spenders["Item ID"].count()
Top_Spenders["Average Purchase Price"] = Spenders["Price"].sum()/Top_Spenders["Purchase Count"]
Top_Spenders["Total Purchase Value"] = Spenders["Price"].sum()

Top_Sort = Top_Spenders.sort_values("Purchase Count", ascending=False)
Top_Sort["Average Purchase Price"] = Top_Sort["Average Purchase Price"].map("${:.2f}".format)
Top_Sort["Total Purchase Value"] = Top_Sort["Total Purchase Value"].map("${:.2f}".format)
Top_Sort.head()
```

## Most Popular Items
* Retrieve the Item ID, Item Name, and Item Price columns
* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
* Create a summary data frame to hold the results
* Sort the purchase count column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame


```python
Item = purchase_df.groupby(["Item ID","Item Name"])

Popular = pd.DataFrame()
Popular["Purchase Count"] = Item["Purchase ID"].count()
Popular["Item Price"] = Item["Price"].sum()/Popular["Purchase Count"]
Popular["Total Purchase Value"] = Item["Price"].sum()
Pop_Sort = Popular.sort_values("Purchase Count", ascending=False)

Pop_Sort["Item Price"] = Pop_Sort["Item Price"].map("${:.2f}".format)
Pop_Sort["Total Purchase Value"] = Pop_Sort["Total Purchase Value"].map("${:.2f}".format)

Pop_Sort.head()
```
